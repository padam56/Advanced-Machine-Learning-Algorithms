```python
# Problem Statement (Classification): To predict whether a product sold by Synergix would seil greater than 1000 units based on factors such as page traffic and unit price.
# #Importing necessary libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv('train_new.csv')
train_df
```

```python
val_df =pd.read_csv('val_new.csv')
val_df
```

```python
test_df = pd.read_csv('test_new.csv')
test_df
```

```python
y_train = train_df['Units_sold>1000']
X_train = train_df.drop('Units_sold>1000', axis=1)

y_validation = val_df['Units_sold>1000']
X_validation = val_df.drop('Units_sold>1000', axis=1)


y_test = test_df['Units_sold>1000']
X_test = test_df.drop('Units_sold>1000', axis=1)
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
# Initialize the base learners
rf = RandomForestClassifier(max_depth=20, min_samples_leaf=10, n_estimators=61, random_state=42, n_jobs=-1)
knn = make_pipeline(StandardScaler(), KNeighborsClassifier(n_neighbors=5))
gb = GradientBoostingClassifier(n_estimators=400, learning_rate=0.3, max_depth=10, random_state=42)

# Create the Voting Classifier with hard voting
voting_clf = VotingClassifier(estimators=[
    ('rf', rf),
    ('knn', knn),
    ('gb', gb)
], voting='hard')
```

```python
# Train the Voting Classifier
voting_clf.fit(X_train, y_train)

# Predict on the training and validation set
y_pred_train = voting_clf.predict(X_train)
y_pred_validation = voting_clf.predict(X_validation)
y_pred_test = voting_clf.predict(X_test)
```

```python
from sklearn.metrics import f1_score

# Calculate F1 score for training and validation data
f1_score_train = f1_score(y_train, y_pred_train)
f1_score_validation = f1_score(y_validation, y_pred_validation)
f1_score_test = f1_score(y_test, y_pred_test)

print(f"F1 Score for Training Data: {f1_score_train}")
print(f"F1 Score for Validation Data: {f1_score_validation}")
print(f"F1 Score for Test Data: {f1_score_test}")
```
